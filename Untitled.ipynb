# QR Code

Pure Python **Quick Response** Code Generator.

QR : 2D Pictographic Code used for Fast Readability. 

Code consist of **Black** Modules arranged in **Square** Pattern on White Background.

```python
import qrcode
img = qrcode.make("https://github.com/KIRANKUMAR7296/")
img.save("GitHub.jpg")
```

# Open Source Computer Vision

Convert Image to NumPy Array Structure.

```python
import cv2
detect = cv2.QRCodeDetector()
encoded_string ,points, QR= detect.detectAndDecode(cv2.imread("GitHub.jpg"))
```
